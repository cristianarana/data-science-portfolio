### Data Transformation

##### Here we are going to generate o group data so we'll make clustering by different groups.

```python
import pandas as pd
from datetime import datetime
```

```python
df_main = pd.read_csv("../data/original/marketing_campaign.csv", sep="\t", header=0)
df_main.head()
```

First we are going to separate to answer "Who is the client?"


```python
age = datetime.now().year - df_main["Year_Birth"]
df_customer_profile = df_main[
    ["ID", "Education", "Marital_Status", "Kidhome", "Teenhome", "Income"]
].assign(Age=age)
df_customer_profile_encoded = pd.get_dummies(
    df_customer_profile,
    columns=["Education", "Marital_Status"],
    drop_first=True
)
df_customer_profile_encoded.to_csv("../data/processed/customer_profile.csv", index=False)
df_customer_profile_encoded.head()
```

Now we are going to separate to identify "How the client makes his purchase?"

```python
df_purchases = df_main[["NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases", "NumWebVisitsMonth"]]
df_purchases.to_csv("../data/processed/purchases.csv", index=False)
df_purchases.head()
```

Here we are going to identify "How the client reacts to marketing campaings?"

```python
AcceptedCmp_Total = df_main[
    ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3',
     'AcceptedCmp4','AcceptedCmp5']
].sum(axis=1)
df_marketing_response = df_main[
    ["NumDealsPurchases", "Response"]
].assign(AcceptedCmp_Total=AcceptedCmp_Total)
df_marketing_response.head()
df_marketing_response.to_csv("../data/processed/marketing_response.csv", index=False)
```

And finally we are going to group to answer "Which is the behavior of the client?"

```python
total_spending = df_main[
    [ "MntWines", "MntFruits", "MntMeatProducts",
    "MntFishProducts", "MntSweetProducts", "MntGoldProds"]
].sum(axis=1)
df_purchase_behavior = df_main[
    ["Recency",  "MntWines", "MntFruits", "MntMeatProducts",
    "MntFishProducts", "MntSweetProducts", "MntGoldProds"]
].assign(Total_Spending=total_spending)
df_purchase_behavior.head()
df_purchase_behavior.to_csv("../data/processed/purchase_behavior.csv", index=False)
```
